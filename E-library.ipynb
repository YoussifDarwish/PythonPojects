```python
class books:
    
    def __init__(self, name, writer, price, details):
        
        self.name = name
        self.writer = writer
        self.price = price
        self.details = details
        
class library:
    def __init__(self):
        self.Books = []
        self.returned_books = []
        self.borrowed_books = []
        
    def add_book(self, name, writer, price, details):
        Book = books(name, writer, price, details)
        self.Books.append(Book)
        print("book added successfully!!")
        
    def remove_book(self, name, writer, price, details):#to remove books from the list (for admin or borrowed books)
        for book in self.Books:
            if name == book.name:
                self.Books.remove(book)
                print("book removed!")
                       
            
    def borrow_book(self):
        while True:
            self.display()
            print("1: borrow")
            print("2: back")
            options = input("option")
            if options == "1":
                name = input("book name: ")
                for book in self.Books:
                    if name == book.name:
                        self.borrowed_books.append(book)
                        self.Books.remove(book)
                        print("book borrowed")   
                        break
                if name != book.name:
                    print("book doesn't exist")
                    
            elif options == "2":
                break
            
        
    
    def return_book(self):
        
        name = input(" book name: ")
        for book in self.borrowed_books:
            if name == book.name:
                self.Books.append(book)
        print("book returned!")
        
        
    
    def display(self):
        for book in self.Books:
            print(f"Book: {book.name}, Writer: {book.writer}, Price: {book.price}, Details: {book.details}")
            
    def display_admins(self):
        for book in self.returned_books:
            print(f"book: {book}")
            
    def menu(self):
        print("1: Home")
        print("2: borrow book")
        print("3: return book")
        print("4: exit")
        print("5: admin")
        
    def admin(self):  #open admins version to add/remove books
        while True:
            print("options:")
            print("1: add book")
            print("2: remove book")
            print("3: display")
            print("4 exit")
            option = input("option: ")
            
            if option == "4":
                print("thank you for using our app!")
                break
                
            elif option == "1":
                name = input("name: ")
                writer = input("writer: ")
                price = input("price: ")
                details = input("details: ")
                self.add_book(name, writer, price, details)
                
            elif option == "2":
                name = input("book name: ")
                self.remove_book(name,writer,price,details)
                
            elif option == "3":
                self.display_admins()
                
               
                    
    def run(self): #users' app
        while True:
            self.menu()
            option = input("option: ")
            if option == "4":
                print("Thank you for using our app!")
                break
                
            elif option == "1":
                self.display()
                
            elif option == "2":
                self.borrow_book()
                
            elif option == "3":
                self.return_book()
            
            elif option == "5":
                self.admin()
                
                
                    
                
                
            
```

```python
app = library()

app.run()
```
